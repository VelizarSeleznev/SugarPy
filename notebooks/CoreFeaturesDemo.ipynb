# SugarPy Core Features Demo

This notebook shows the main classroom flow in SugarPy:
- Core math behavior
- Core code behavior
- Optional chemistry helpers

Run top-to-bottom.


## 1) Setup and quick API map

This cell confirms the main public helpers are available.


```python
from sugarpy import (
    balance_equation,
    render_stoichiometry,
    load_catalog,
    function_builder,
    balance_widget,
)
from sugarpy.math_cell import render_math_cell

print('SugarPy imports: OK')
print('Catalog ids:', [f.id for f in load_catalog()])
```

## 2) Math core: expression, symbolic, equation, assignment

SugarPy math parser rules:
- `=` means equation
- `:=` means assignment
- `^` is supported as power


```python
samples = {
    'numeric_expression': render_math_cell('2 + 2'),
    'symbolic_expression': render_math_cell('x + 2'),
    'equation': render_math_cell('x^2 = 2'),
    'assignment': render_math_cell('a := 5'),
}

for name, payload in samples.items():
    print(f'\n{name}')
    print('  ok   :', payload['ok'])
    print('  kind :', payload['kind'])
    print('  value:', payload['value'])
    print('  error:', payload['error'])
```

## 3) Trig mode demo (degrees vs radians)


```python
deg = render_math_cell('sin(30)', mode='deg')
rad = render_math_cell('sin(30)', mode='rad')

print('deg value:', deg['value'])
print('rad value:', rad['value'])
```

## 4) Code + Math collaboration

In SugarPy UI, Code cells and Math cells share the namespace.
Here we simulate that by giving a custom `user_ns` to the math parser.


```python
from sugarpy.math_parser import parse_sympy_expression

user_ns = {}

def f(x):
    return x**2

user_ns['f'] = f
user_ns['a'] = 5

expr = parse_sympy_expression('f(3) + 2a', mode='deg', user_ns=user_ns)
print('f(3) + 2a ->', expr)
```

## 5) Optional chemistry


```python
print(balance_equation('Fe + O2 -> Fe2O3'))
print(balance_equation('C3H8 + O2 -> CO2 + H2O'))

table = render_stoichiometry('H2 + O2 -> H2O', {'H2': {'n': 2}})
print('stoich ok:', table['ok'])
print('balanced :', table['balanced'])
print('species  :', [row['name'] for row in table['species']])
```

## 6) Final self-check

If this cell passes, the core demo is healthy.


```python
assert render_math_cell('2 + 2')['ok']
assert render_math_cell('x^2 = 2')['kind'] == 'equation'
assert render_math_cell('a := 5')['kind'] == 'assignment'
assert balance_equation('H2 + O2 -> H2O') == '2H2 + O2 -> 2H2O'
assert render_stoichiometry('H2 + O2 -> H2O', {'H2': {'n': 2}})['ok']

print('All checks passed.')
```
